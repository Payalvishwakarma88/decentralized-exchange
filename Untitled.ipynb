# **Kidney Disease Analysis**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

#data set
df = pd.read_csv("/content/kidney_disease_dataset.csv")
df
x = df[["Age","Creatinine_Level","BUN","Diabetes","Hypertension","GFR","Urine_Output","CKD_Status","Dialysis_Needed"]]
y = df["Dialysis_Needed"]

#spliting the data
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2)

#algorithm
model = LinearRegression()

#training the model
model.fit(x_train, y_train)

#testing the model
model.fit(x_train, y_train)

#testing
y_pred = model.predict(x_test)
MAE = mean_absolute_error(y_test, y_pred)
print(MAE)

#testing
y_pred = model.predict(x_test)
MAE = mean_absolute_error(y_test, y_pred)
print("mean absolute error:", MAE)
accu = r2_score(y_test, y_pred)
print("accuracy score:", accu*100)
```

polynomial **regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score


df = pd.read_csv("/content/kidney_disease_dataset.csv")
df
x = df[["Age"]]
y = df["Dialysis_Needed"]

poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_pred = model.predict(x_poly)


plt.scatter(x,y,color="blue",label="actual data")
plt.plot(x,y_pred,color="red",label="polnomialfit")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```
